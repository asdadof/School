## Oppgave 1
Tegn 100 tilfeldige punkter, der b친de _x_- og _y_-verdiene ligger i intervallet $[0,100>$.

<details>
    <summary style="color:green;">游띋 Klikk her for et hint</summary>
    Vi trekker et tilfeldig heltall med funksjonen <code>randint(start,slutt)</code>.

```python
import random as rd
import matplotlib.pyplot as plt 

x_liste = []
y_liste = []

for i in range(100):
    x_liste.append(rd.randint(1,99))
    y_liste.append(rd.randint(1,99))


plt.plot(x_liste, y_liste, "o")
plt.show()

```

## Oppgave 2
Tegn 500 tilfeldige linjer. Hver linje skal trekkes mellom to punkter, der _x_-verdien skal ligge i intervallet $[0, 320>$ og _y_-verdien skal ligge i intervallet $[0,200>$.


<details>
    <summary style="color:green;">游띋 Klikk her for et hint</summary>
    Her m친 du tegne en og en linje inne i en l칮kke.
</details>

```python
import random as rd
import matplotlib.pyplot as plt 

x_liste = []
y_liste = []

for i in range(500):
    x_liste.append(rd.randint(0,320))
    y_liste.append(rd.randint(1,200))


plt.plot(x_liste, y_liste, "o-")
plt.show()
```
