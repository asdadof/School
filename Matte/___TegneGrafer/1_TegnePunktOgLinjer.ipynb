# Tegne punkt og linjer

Når vi ønsker å tengne punkt, linjer, grafer og så videre, må vi importere et bibliotek som er spesiallaget for dette.

Vi legger da til følgende linje i starten av programmet:

```python
import matplotlib.pyplot as plt
```

Her bruker vi stikkordet <code>as</code> til å gi biblioteket et kortere navn, et alias. I koden videre kan vi dermed bruke navnet <code>plt</code> når vi ønsker å bruke biblioteket.

## Eksempel 1

> Tegn punktene (3, 2), (-2,1), (-1,-4) og (4, -3) i samme koordinatsystem.

Vi starter med å definere to lister. En liste med punktenes <i>x</i>-verdier, og en liste med punktenes <i>y</i>-verdier.

```python
x_liste = [ 3, -2, -1,  4]
y_liste = [ 2,  1, -4, -3]
```

Over ser vi de fire punkene representert ved fire kolonner. 

Når punktene er lagt i listene, bruker vi metoden <code>.plot()</code> til å klargjøre plottet.

Metoden tar tre parametre, de to listene og instillinger for hvordan skjermbildet skal se ut. <code>'o'</code> betyr her at vi ønsker at punktene skal tegnes med prikker. Vi kunne her brukt <code>'x'</code> dersom vi i stedet ønsket kryss.

```python
plt.plot( x_liste, y_liste, 'o' )
plt.show()
```

Metoden <code>.show()</code> viser det ferdige bildet.

## Eksempel 2

> Tegn punkene (0,0), (1,3), (3, -1), (5,2) og (7,1). Trekk linje mellom punktene.

Vi definerer først listene.

```python
x_liste = [ 0,  1,  3,  5,  7]
y_liste = [ 1,  3, -1,  2,  1]
```

Vi bruker denne gang <code>'o-'</code> som innstilling for tegning.

```python
plt.plot( x_liste, y_liste, 'o-' )
plt.show()
```

Vi kan også stille inn fargen som brukes. <code>r</code> i <code>'o-r'</code> betyr at vi vil bruke rød. 

```python
plt.plot( x_liste, y_liste, 'o-r' )
plt.show()
```

## Eksempel 3
Vi kan tegne flere grafer i samme bilde.
> Trekk en rød linje mellom punktene (1,3), (4,1), (7, 4), (8, 1) og (12, 7), og en grønn linje mellom punktene (1, 2), (5, -2), (7, 2) og (12, 5).

```python
x_liste1 = [ 1,  4,  7,  8, 12]
y_liste1 = [ 3,  1,  4,  1,  7]

x_liste2 = [ 1,  5,  7, 12]
y_liste2 = [ 2, -2,  2,  5]

plt.plot( x_liste1, y_liste1, 'r' )
plt.plot( x_liste2, y_liste2, 'g' )

plt.show()
```
